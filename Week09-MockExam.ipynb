```python
LarcenyTs.loc[(LarcenyTs.Year==2014), 'CrimeDate'].value_counts().values
Larceny_2014['Date']=LarcenyTs.loc[(LarcenyTs.Year==2014), 'CrimeDate'].value_counts().index
Larceny_2014['Occurence']=LarcenyTs.loc[(LarcenyTs.Year==2014), 'CrimeDate'].value_counts().values
Larceny_2015['Date']=LarcenyTs.loc[(LarcenyTs.Year==2015), 'CrimeDate'].value_counts().index
Larceny_2015['Occurence']=LarcenyTs.loc[(LarcenyTs.Year==2015), 'CrimeDate'].value_counts().values
Larceny_2016['Date']=LarcenyTs.loc[(LarcenyTs.Year==2016), 'CrimeDate'].value_counts().index
Larceny_2016['Occurence']=LarcenyTs.loc[(LarcenyTs.Year==2016), 'CrimeDate'].value_counts().values
Larceny_df = Larceny_2014.append(Larceny_2015, ignore_index=True)
Larceny_df = Larceny_df.sort_values(by=['Date'])
Larceny_df = Larceny_df.reset_index(drop=True)
Larceny_df
```

```python
##### Step 0 - We will use this function in step 3 to get the data into the right shape
def prep_data(datain, time_step):
    # 1. y-array  
    # First, create an array with indices for y elements based on the chosen time_step
    y_indices = np.arange(start=time_step, stop=len(datain), step=time_step)
    # Create y array based on the above indices 
    y_tmp = datain[y_indices]
    
    # 2. X-array  
    # We want to have the same number of rows for X as we do for y
    rows_X = len(y_tmp)
    # Since the last element in y_tmp may not be the last element of the datain, 
    # let's ensure that X array stops with the last y
    X_tmp = datain[range(time_step*rows_X)]
    # Now take this array and reshape it into the desired shape
    X_tmp = np.reshape(X_tmp, (rows_X, time_step, 1))
    return X_tmp, y_tmp
```

```python
x=Larceny_df[['Occurence']]
scaler = MinMaxScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
x_train, y_train = prep_data(x_scaled, 10)
```

```python
model1 = tf.keras.models.Sequential()
model1.add(tf.keras.layers.SimpleRNN(units=50, input_shape=(10,1), activation="relu",return_sequences=True))
model1.add(tf.keras.layers.SimpleRNN(units=50, activation="relu",return_sequences=True))
model1.add(tf.keras.layers.SimpleRNN(units=50, activation="relu",return_sequences=True))
model1.add(tf.keras.layers.SimpleRNN(units=50, activation="relu"))
model1.add(tf.keras.layers.Dense(1))
model1.compile(loss='mean_squared_error', optimizer='rmsprop')
model1.summary()
```

```python
history1 = model1.fit(x=x_train, y=y_train, batch_size=32, epochs=100)
```
